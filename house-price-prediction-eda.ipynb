```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'C:\Users\Administrator\Desktop\Exploratory Data Analysis\House Price Prediction\house-price-prediction-eda\dataset\housing_price_dataset.csv')
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Price'], kde=True, color='teal')
plt.title('Distribution of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
num_cols = ['Bedrooms', 'Bathrooms', 'SquareFeet', 'YearBuilt']
for col in num_cols:
    plt.figure(figsize=(7, 5))
    sns.scatterplot(data=df, x=col, y='Price', alpha=0.5)
    plt.title(f'Price vs {col}')
    plt.xlabel(col)
    plt.ylabel('Price')
    plt.show()
```

```python
cat_cols = ['Neighborhood']
for col in cat_cols:
    plt.figure(figsize=(10, 5))
    sns.boxplot(data=df, x=col, y='Price')
    plt.title(f'Price by {col}')
    plt.xlabel(col)
    plt.ylabel('Price')
    plt.xticks(rotation=45)
    plt.show()
```

```python
for col in cat_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=col, palette='pastel')
    plt.title(f'Count of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df.sort_values('YearBuilt'), x='YearBuilt', y='Price')
plt.title('Average Price by Year Built')
plt.xlabel('Year Built')
plt.ylabel('Price')
plt.show()
```
